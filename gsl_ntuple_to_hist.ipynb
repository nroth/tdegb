```python
import numpy as np
%matplotlib nbagg
import matplotlib
import matplotlib.pyplot as plt
from math import exp, sqrt, fabs, pow, log, log10
from scipy.integrate import simps, quad

gaussian_hist = np.loadtxt("gsl_hist_1d_m_r.hist")
left_edges = gaussian_hist[::,0]
right_edges = gaussian_hist[::,1]
bin_centers = 0.5 * (left_edges + right_edges) # assumes uniform spacing
widths = right_edges - left_edges
counts = gaussian_hist[::,2]


print("Ready")

```

```python
plt.close()

print(counts.sum())
plt.bar(bin_centers,counts, width = widths, align='center',edgecolor='k',color='white')

#plt.xlim(0.,5.)


```

```python
plt.close()

filename = "gsl_hist_2d_m_r_z.hist"
m_r_z_hist_counts = np.loadtxt(filename,skiprows=2,usecols=2) # fix the name, 2D

z_grid = np.loadtxt(filename,max_rows=1) # edges
m_r_grid = np.loadtxt(filename,skiprows=1, max_rows=1) #edges

m_r_z_hist_counts = m_r_z_hist_counts.reshape(len(z_grid) - 1 ,len(m_r_grid) - 1) # centers

z_mesh,m_r_mesh = np.meshgrid(z_grid,m_r_grid,indexing='ij')

# flat or gauraud
#plt.pcolormesh(m_r_mesh,z_mesh, m_r_z_hist_counts,shading='gauraud',cmap="jet")                                                                           
plt.pcolormesh(z_mesh,m_r_mesh, m_r_z_hist_counts,shading='gauraud',cmap="jet")  


cbar = plt.colorbar()                                                                          
max_count = np.amax(m_r_z_hist_counts)                                                       
min_count = np.amin(m_r_z_hist_counts)                                                       
plt.clim(min_count,max_count)  
#plt.clim(0,1200)
                                                      
                                                                                           
cbar.ax.tick_params(labelsize=18)                                                          
cbar.set_label(r'log density (code units)',fontsize=16)                                    
                                                                                           
                                          
plt.xlabel("$z$",fontsize=16)                                
plt.ylabel("m_r",fontsize=16)                                                          
                                                                                               

print("Total counts is %e" % m_r_z_hist_counts.sum())


                                                                                           
                                            
#savefig("blah.png")
```
