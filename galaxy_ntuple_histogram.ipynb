```python
import numpy as np
%matplotlib nbagg
import matplotlib
import matplotlib.pyplot as plt
import tdegbviz
```

```python
plt.close()

data_path = 'results/third_production_runs/v7/'
save_path = data_path

b = tdegbviz.Binner(data_path= data_path,save_path = save_path)
print('ready')



```

```python
hist_type = 'gals'
axis = 'log_mbh_sigma'

b.set_data_path('results/third_production_runs/v6/')
b.set_save_path(data_path)

b.set_hist_type(hist_type)

plt.close()
hist = b.plot_1d_histogram(axis)
plt.setp(hist,edgecolor = 'k',linewidth=1) # can change other parameters here

plt.xlabel(r'$log_{10}(M_{\rm BH}/M_\odot)$')

#plt.xlim(3,9)
#plt.ylim(0,100000)

#b.save_current_histogram()
```

```python
hist_type = 'vol_disrupt'
contour_type = 'flares'
a1 = 'UminusR'
a2 = 'mstar_mendel'
b.set_data_path('results/third_production_runs/v7/')
cmap_name = 'gist_stern'
truncate_colors_max = 1.


########

b.set_save_path(data_path)
b.set_hist_type(hist_type)

# make the pcolor plot
plt.close()
b.read_2d_histogram(axis1 = a1,axis2 = a2,transpose_default = 0)
hist2d, cbar = b.plot_2d_histogram(cmap_name = cmap_name, cm_maxval = truncate_colors_max)

# make the contour plot
b.set_hist_type(contour_type)
b.read_2d_histogram(axis1 = a1,axis2 = a2,transpose_default = 0)
conts = b.plot_2d_contours()

# change contour properties
for contour in conts.collections:
    plt.setp(contour,linewidth = 2,color='w')
    

# add contour labels and set their properites
fmt = {}
strs = ['10%', '30%', '50%', '70%', '90%']
for l, s in zip(conts.levels, strs):
    fmt[l] = s
plt.gca().clabel(conts, conts.levels[0:3], inline=True, fmt=fmt, fontsize=10, colors = 'w')

# change colorbar properties
cbar.set_label('new label')
cbar.ax.tick_params(labelsize=16)

# add data points
host_data = np.loadtxt("ztf_tde_host_data_popsyn.txt",skiprows=1, usecols=(1,2,3,4))
for i in range(len(host_data)):
    plt.plot(host_data[i,0],host_data[i,1],'*',markersize=15,c = plt.cm.jet(195),markeredgecolor ='k')


# change axes labels or histogram limits
plt.xlabel(r"host $M_\ast$")
plt.ylabel(r"host $u - r$")
#plt.xlim(8.5,12)
#plt.clim(0,20000)

#b.save_current_histogram()

```
