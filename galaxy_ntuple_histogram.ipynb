```python
import numpy as np
%matplotlib nbagg
import matplotlib
import matplotlib.pyplot as plt
import tdegbviz

b = tdegbviz.Binner()
print('ready')
```

```python
hist_type = 'flares'
axis = 'log_mbh_sigma'
data_path = 'results/fourth_production_local/v1/'

b.set_data_path(data_path)
b.set_save_path(data_path)
b.set_hist_type(hist_type)

plt.close()
hist = b.plot_1d_histogram(axis) 
plt.setp(hist,edgecolor = 'k') # can change other parameters here

plt.xlabel(r'$\log_{10}(M_{\rm BH}/M_\odot)$')
#plt.xlabel(r'$T_{\rm bb}$ fit (K)')
plt.ylabel(r'ZTF detections / yr / bin')

#plt.xlim(3,9)
#plt.ylim(0,100000)

b.save_current_histogram()



```

```python
hist_type = 'gals'
contour_type = 'flares'
a1 = 'UminusR'
a2 = 'log_mstar_mendel'
b.set_data_path('results/fourth_production_local/v1/')
#b.set_data_path('./')
cmap_name = 'YlGnBu'
truncate_colors_max = 0.7
transpose = 0

contour_levels = [0.1,0.3,0.5,0.7,0.9]


########

b.set_save_path(data_path)
b.set_hist_type(hist_type)

# make the pcolor plot
plt.close()
b.read_2d_histogram(axis1 = a1,axis2 = a2,transpose_default = transpose)
hist2d, cbar = b.plot_2d_histogram(cmap_name = cmap_name, cm_maxval = truncate_colors_max)

# make the contour plot
b.set_hist_type(contour_type)
b.read_2d_histogram(axis1 = a1,axis2 = a2,transpose_default = transpose)
conts = b.plot_2d_contours(contour_levels)

# change contour properties
for contour in conts.collections:
    plt.setp(contour,linewidth = 2,color='k')
    
# add contour labels and set their properites
#fmt = {}
#strs = ['10%', '30%', '50%', '70%', '90%']
#for l, s in zip(conts.levels, strs):
#    fmt[l] = s
#plt.gca().clabel(conts, conts.levels[0:3], inline=True, fmt=fmt, fontsize=10, colors = 'w')

# change colorbar properties
cbar.set_label('galaxies / bin (full catalogue)')
cbar.ax.tick_params(labelsize=16)

# add data points
host_data = np.loadtxt("ztf_tde_host_data_popsyn.txt",skiprows=1, usecols=(1,2,3,4))
for i in range(len(host_data)):
    plt.plot(host_data[i,0],host_data[i,1],'*',markersize=15,c = plt.cm.jet(195),markeredgecolor ='k')


# change axes labels or histogram limits
plt.xlabel(r"host $M_\ast$")
plt.ylabel(r"host $u - r$")
#plt.xlim(8.5,12)
#plt.clim(0,20000)

b.save_current_histogram()

```
